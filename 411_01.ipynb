```
import pandas as pd #Importē moduli
df = pd.read_csv('EventLogs.csv') #Nolasa iegūto event log failu
total = len(df.index) #Iegūst kopējo rindu skaitu
Error = df.drop(['Information', 'Warning','Critical']) #Izveido data frame tikai no Erroru eventiem
Information = df.drop(['Error', 'Warning','Critical']) #Izveido data frame tikai no Informācijas eventiem
Warning = df.drop(['Error', 'Information','Critical']) #Izveido data frame tikai no Warning eventiem
Critical = df.drop(['Error', 'Information','Warning']) #Izveido data frame tikai no Critical eventiem

my_explode=(1.5,0.0,0.1,0.5)
data = pd.DataFrame({'Events':[len(Error.index),len(Information.index),len(Warning.index),len(Critical.index)]}, #izmantojot len() iegūst individuālu eventu skaitu lai izveidotu grafika data frame
                    index = ['Error', 'Information','Warning','Critical'])
plot = data.plot.pie(y='Events',figsize=(10, 10),autopct='%1.1f%%',explode=my_explode) #Izveido noteikta izmēra grafiku ar procentiem


```
